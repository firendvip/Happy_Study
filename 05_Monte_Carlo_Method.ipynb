# 醉汉随机走

```python
import random  # 随机函数库
import math  # 数学库
```

```python
def random_walk(n):  # 定义随机走的函数
    x,y = 0,0
    for i in range(n):  # 
        dx,dy = random.choice([(0,1),(0,-1),(1,0),(-1,0)])  # 坐标 上下左右 随机选择一次
        x += dx   # 每次累加给到X  
        y += dy  # 每次累加给到Y  
    return(x,y)
```

```python
count = 0   # 计数初始=0
walk_times = 1000  # 走动的次数
walk_steps = 200  # 走动的步数
home_distance_max = 20  # 走动的最远距离
```

```python
for i in range(walk_times):  # 
    one_walk = random_walk(walk_steps) # 每次走完的最周坐标给到 one_walk
    # 最远距离为distance 斜边距离为两直角边长平方和的开方，.sqrt是 求平方根
    distance = math.sqrt((int(one_walk[0])**2+int(one_walk[1])**2))  
    if distance >= home_distance_max:  # 如果距离大于最大步数
        count += 1   # 计数+1
    print(f'coordinate:{one_walk},distance:{distance}')  # 输出坐标和距离
ratio = count/walk_times  #  超出最远距离的比例
print(f'ratio:{ratio}')  # 打印比例
```

# 赌徒输赢概率

```python
import random
import matplotlib   # 画图的库
import matplotlib.pyplot as plt  # 画图的库
```

```python
def give_the_number():   # 获取一个随机数的函数
    number = random.randint(1,100)  # 随机获得1到100的整数
    if number <= 51:    # 获取的数<= 51 返回false 判为输
        return False
    else:
        return True   #  1-49 为赢 胜率 49%
```

```python
def a_fool(funds,wager,max_count):      #  funds资金，基金，wager下注，game_count游戏次数
    game_count= 0
    x=[]
    y=[]
    while game_count < max_count:   # 
        if give_the_number():
            funds+=wager
            x.append(game_count)
            y.append(funds)
        else:
            funds-= wager
            x.append(game_count)
            y.append(funds)
        game_count+=1
    if funds<0:
        funds='Pennyless'
    plt.plot(x,y)
```

```python
n=0
funds=1000000     # 资金有100万
wager=100000    #  每次下注10万
max_count=20000  # 下注2万次
while n <100:   # 模拟100个人
    a_fool(funds,wager,max_count)
    n+=1
plt.xlabel('How many games does the fool played')
plt.ylabel('Funds')
plt.show()
```
