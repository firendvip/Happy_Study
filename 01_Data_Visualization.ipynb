```python
import matplotlib.pyplot as plt

x = [11,22,3,4,3,6,7,8,9,10,11,12]
y = [1,22,3,4,33,6,7,28,9,22,77,12]
x2 = [1,2,3,4,5,45,7,8,9,10,11,12]
y2 = [11,33,44,5,52,6,7,11,9,22,33,12]

plt.plot(x,y,label = 'Company A') # x,y里的数值个数要一样
plt.plot(x2,y2,label = 'Company B') # x,y里的数值个数要一样
plt.title('Seles of Company\nCompany A and Company B') # 图表标题 可以\n换行
plt.xlabel('Month') # x 轴的标签 
plt.ylabel('Sales\n(yuan)') # y 轴的标签
plt.legend() # 给图像加上图例 
plt.show()
```

### plt.hist()  直方图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
low = 1
high = 101
```

```python
City_A_Age = list(np.random.randint(low,high,size=60))  # low到high-1的范围内 生成 size个整数
print(City_A_Age,type(City_A_Age))
```

```python
bins = list(range(0,111,10))
print(bins)
```

```python
# bins为数据区间切分，histtype='bar'图表类型，rwidth=0.8数据图宽度，一般为0.8
plt.hist(City_A_Age,bins,histtype='bar',rwidth=0.8,label='City_A_Age') 
plt.title('The City A Age Hist') # 图表标题 可以\n换行
plt.xlabel('Age') # x 轴的标签 
plt.ylabel('Num') # y 轴的标签
plt.legend() # 给图像加上图例  
plt.show()
```

### plt.scatter()散点图

```python
import matplotlib.pyplot as plt
import numpy as np
x = list(range(1,101))
y = list(np.random.randint(1,11,size=100)) 
# marker='*' 以星星显示，默认为圆点，,s=100，散点的大小
plt.scatter(x,y,label='scatter_plot',color='b',marker='*',s=100) 
plt.title('A Scatter_plot') # 图表标题 可以\n换行
plt.legend() # 给图像加上图例 
plt.xlabel('x') # x 轴的标签 
plt.ylabel('y') # y 轴的标签
plt.show()
```

### plt.stackplot()堆栈图

```python
import matplotlib.pyplot as plt
import numpy as np

month = list(range(1,13))
A = list(np.random.randint(10,25,size=12)) 
B = list(np.random.randint(20,30,size=12)) 
C = list(np.random.randint(100,300,size=12)) 
D = list(np.random.randint(50,90,size=12)) 

plt.plot([],[],color='m',label='Company A',linewidth=6)
plt.plot([],[],color='c',label='Company B',linewidth=6)
plt.plot([],[],color='r',label='Company C',linewidth=6)
plt.plot([],[],color='k',label='Company D',linewidth=6)

plt.stackplot(month,A,B,C,D,colors=['m','c','r','k'])

plt.xlabel('month') # x 轴的标签 
plt.ylabel('sales') # y 轴的标签
plt.title('A B C D sales') # 图表标题 可以\n换行
plt.legend() # 给图像加上图例 
plt.show()
```

### plt.pie() 饼状图

```python
import matplotlib.pyplot as plt
import numpy as np

sales = [1.8,2,3.5,7.8]
company_names = ['Company A','Company B','Company C','Company D']
colors_1 = ['azure','lavender','pink','aqua']

plt.pie(sales,# 数据源
        labels=company_names, # 模块标题
        colors = colors_1, # 模块颜色
        startangle=90, # 从90度开始切割
        shadow=True, # 有阴影
        autopct='%1.2f%%',  # 2 表示保留小数后两位
        explode=(0.15,0,0,0)) # 0.15的程度爆裂出 位置一的模块 

plt.legend()
plt.show()
```

### 多个数据画图叠加

```python
from matplotlib import pyplot as plt
x = list(range(1,11))
y = list(range(1,20,2))
z = list(range(2,21,2))
o = list(range(3,13))
print(len(x),len(y),len(z),len(o))
plt.plot(x,y,'o') # 以圆点做线
plt.plot(x,z,'*') # 以星星做线
plt.plot(x,o,'x') # 以X做线
plt.xlabel('x') # x 轴的标签 
plt.ylabel('y') # y 轴的标签
plt.title('three Plot') # 图表标题 可以\n换行
plt.legend(['this is y','this is z','this is o']) # 给图像加上图例 
plt.show()
```

### 世界国家人口增长

```python
from matplotlib import pyplot as plt
import pandas as pd
path = r'C:\Users\10746\HappyStudy\Data_Warehouse\DataVisualization\countries.csv\countries.csv'
f = open(path,encoding='utf-8')  # 路径带有中文的时候 要先打开才能用pd读取
data2 = pd.read_csv(f)
```

```python
data2.year.iloc[0]  # year列0行的数据
```

```python
China_filter = data2.country == 'China'  # 取data2里country值为China的数据 比较后的结果为 bool值
China_data = data2[China_filter]  # 取 data2里country值为China的数据

US_filter = data2.country == 'United States'
US_data = data2[US_filter]

JP_filter = data2.country == 'Japan'
JP_data = data2[JP_filter]
```

```python
plt.plot(China_data.year,China_data.population/10**8)  # 除于10的8次方即 以亿 为单位
plt.plot(US_data.year,US_data.population/10**8)
plt.plot(JP_data.year,JP_data.population/10**8)
plt.xlabel('Year') # x 轴的标签 
plt.ylabel('Population') # y 轴的标签
plt.title('Country Population') # 图表标题 可以\n换行
plt.legend(['China','United States','Japan']) # 给图像加上图例 
plt.show()
```

```python
plt.plot(China_data.year,China_data.population/China_data.population.iloc[0]) # 以1950年作为基数的增长率作为Y轴数据
plt.plot(US_data.year,US_data.population/US_data.population.iloc[0])
plt.plot(JP_data.year,JP_data.population/JP_data.population.iloc[0])
plt.xlabel('Year') # x 轴的标签 
plt.ylabel('Population') # y 轴的标签
plt.title('Country Population Growth Rate') # 图表标题 可以\n换行
plt.legend(['China','United States','Japan']) # 给图像加上图例 
plt.show()
```
